```python
one = pd.read_excel('../../spreadsheets/informe_20210602.ods', engine="odf", sheet_name=3)
two = pd.read_excel('../../spreadsheets/informe_20210602.ods', engine="odf", sheet_name=4)
```

- one: data for at least one dose
- two: data for two doses 

```python
one = one.rename(columns={"Unnamed: 0": "ccaa"})
two = two.rename(columns={"Unnamed: 0": "ccaa"})

one = one[one['ccaa'] != 'Fuerzas Armadas'].dropna()
two = two[two['ccaa'] != 'Fuerzas Armadas'].dropna()
```

# Preprocessing

```python
# simplify header names
one.columns = [col.replace('Personas con al menos 1 dosis', 'vax1_').replace('años', '').replace('Población INE', 'pop_').replace(' ','').strip() for col in one.columns]
```

```python
age2 = age.copy()
age2['70+'] = age2['80+'] + age2['70-79']
age2['50-69'] = age2['60-69'] + age2['50-59']
```

```python
age_regrouped = age2[['70+', '50-69', '25-49', '18-24', '16-17']]
```

```python
age_regrouped.div(age_regrouped.sum(axis=1), axis=0)
```

```python
age_regrouped.div(age_regrouped.sum(axis=1), axis=0).style.bar(subset=["70+",], color='lightgreen')
```

```python
# percentage point difference from national level
a = age_regrouped.div(age_regrouped.sum(axis=1), axis=0)
a = a - [0.169302,0.308257,0.416302,0.082490,0.023649]


a = a.sort_values(by=['70+'], ascending=False)
def color_negative_red(val):
    """
    Takes a scalar and returns a string with
    the css property `'color: red'` for negative
    strings, black otherwise.
    """
    color = 'hotpink' if val < 0 else 'black'
    return 'color: %s' % color
s = a.style.applymap(color_negative_red).format("{:.2%}")
s
```
